```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy import gradient, linspace
from scipy import odr
from math import e

config = {
'axes.spines.right': False,
'axes.spines.top': False,
'axes.edgecolor': '.4',
'axes.labelcolor': '.0',
'axes.titlesize': 'large',
'axes.labelsize': 'medium',
'figure.autolayout': True,
'figure.figsize': (4.5, 3.5),
'font.family': ['serif'],
'font.size': 10.0,
'grid.linestyle': '--',
'legend.facecolor': '.9',
'legend.frameon': True,
'savefig.transparent': True,
'text.color': '.0',
'xtick.labelsize': 'small',
'ytick.labelsize': 'small',
}
plt.style.use(['seaborn-whitegrid', 'seaborn-paper', 'seaborn-muted', config])

paginas = ['A1H1', 'A1H2', 'A1H3', 'A1H4', 'A1H5', 'A3H1', 'A4H1', 'A5H1']

for i in paginas: 
    #pegando dados
    dados = pd.read_excel('dados.xlsx', sheet_name=i)
    
    t = dados['t [s]']
    r = dados['raio r(t) [cm]']
    dt = dados['incerteza de t [s]'] 
    dr = dados['incerteza de r(t) [cm]']
    
    #fazendo regressão 
    reg = odr.RealData(t, r)
    odreg = odr.ODR(reg, odr.exponential)
    ans = odreg.run()
    beta, beta1 = ans.beta
    incbeta, incbeta1 = ans.sd_beta
    x_regressao = linspace(min(t), max(t), len(t))
    y_regressao = beta + e**(beta1 * x_regressao)
    
    #fazendo derivada com gradiente 
    rotulo = "Raio por tempo"
    dev_dr = gradient(r)
    dev_dt = gradient(t)
    derivada = dev_dr / dev_dt
    print(f'A velocidade inicial no gráfico {i} é', max(derivada))
    
    #plotando dados e derivada com gradiente
    plt.plot(linspace(min(t), max(t), len(derivada)), derivada, label="Derivada de primeira ordem no ponto")
    plt.errorbar(x = t, y = r, xerr = dt, yerr = dr, label = rotulo, fmt='.-k')
    plt.xlabel('Tempo [s]')
    plt.ylabel('Raio [cm]')
    plt.legend()
    plt.title("Variação do raio da circunferência em relação ao tempo")
    plt.savefig(f'grafico-derivada-metodo-fit-exponencial-{i}.jpg', dpi=2000, bbox_inches= 'tight') 
    plt.close()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy import gradient, linspace
from scipy import odr
from math import e
import math
config = {
'axes.spines.right': False,
'axes.spines.top': False,
'axes.edgecolor': '.4',
'axes.labelcolor': '.0',
'axes.titlesize': 'large',
'axes.labelsize': 'medium',
'figure.autolayout': True,
'figure.figsize': (4.5, 3.5),
'font.family': ['serif'],
'font.size': 10.0,
'grid.linestyle': '--',
'legend.facecolor': '.9',
'legend.frameon': True,
'savefig.transparent': True,
'text.color': '.0',
'xtick.labelsize': 'small',
'ytick.labelsize': 'small',
}
plt.style.use(['seaborn-whitegrid', 'seaborn-paper', 'seaborn-muted', config])

paginas = ['A1H1', 'A1H2', 'A1H3', 'A1H4', 'A1H5', 'A3H1', 'A4H1', 'A5H1']

def log(B, x):
    return np.log(B[0] * x) + B[1] 
modelo_log = odr.Model(log)



for i in paginas: 
    #pegando dados
    dados = pd.read_excel('dados.xlsx', sheet_name=i)
    
    t = dados['t [s]']
    r = dados['raio r(t) [cm]']
    dt = dados['incerteza de t [s]'] 
    dr = dados['incerteza de r(t) [cm]']
    
    #fazendo regressão
    min_nonzero_r = np.min(r.to_numpy()[np.nonzero(r.to_numpy())])
    min_nonzero_t = np.min(t.to_numpy()[np.nonzero(t.to_numpy())])
    min_nonzero_dt = np.min(dt.to_numpy()[np.nonzero(dt.to_numpy())])
    min_nonzero_dr = np.min(dr.to_numpy()[np.nonzero(dr.to_numpy())])
    for j in range(len(r)):
        if r[j] == 0:
            r[j] = min_nonzero_r
        if dr[j] == 0:
            dr[j] = min_nonzero_dr
    for k in range(len(t)):
        if t[k] == 0:
            t[k] = min_nonzero_t
        if dt[k] == 0:
            dt[k] = min_nonzero_dt
            
    print(r, t)
    reg = odr.RealData(t, r, sx=dt, sy=dr)
    odreg = odr.ODR(reg, modelo_log, beta0=[.1, .1])
    ans = odreg.run()
    beta0, beta1 = ans.beta
    incbeta0, incbeta1 = ans.sd_beta
    x_regressao = linspace(min(t), max(t), len(t))
    y_regressao = np.log(beta0 * x_regressao) + beta1
    plt.plot(x_regressao, y_regressao, label = f"Regressão logarítmica do tipo $ln({beta0:3.2f} \pm {incbeta0:.2f}) + {beta1:3.2f} \pm {incbeta1:3.2f}$")
    
    #plotando dados e derivada com regressao
    plt.errorbar(x = t, y = r, xerr = dt, yerr = dr, label = rotulo, fmt='.-k')
    plt.xlabel('Tempo [s]')
    plt.ylabel('Raio [cm]')
    plt.legend()
    plt.title("Variação do raio da circunferência em relação ao tempo")
    plt.savefig(f'grafico-derivada-metodo-fit-logaritmica-{i}.jpg', dpi=2000, bbox_inches= 'tight') 
    plt.close()
```
